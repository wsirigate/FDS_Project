```python
df = df.reset_index()
del df["index"]
# del df["level_0"]
```

```python
df.tail(10)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
year_dept = df.iloc[:,3:4]
year_dept["key"] = year_dept.index
```

```python
dummy = pd.get_dummies(df.iloc[:, :3])
```

```python
dummy["key"] = dummy.index
```

```python
# merge
dummy = dummy.merge(year_dept, left_on="key", right_on="key")
```

```python
del dummy["key"]
dummy.head(10)
```

```python
# Elbow method ไว้ดูว่าต้องเลือกค่า k ที่เท่าไหร่่
K = range(2, 15)
sse = []
for k in K:
    model=KMeans(n_clusters=k)
    model = model.fit(dummy)
    sse.append(model.inertia_)

plt.plot(K, sse, 'rx-')

# ค่าที่เริ่ม slove คือค่า k ที่เริ่มจะเหมาะสม
```

```python
# แบ่งกลุ่มข้อมูล

# instantiate
kmean_model = KMeans(n_clusters=6)

# fit
kmean_model.fit(dummy)
```

```python
kmean_model.labels_
```

```python
cluster = kmean_model.labels_
```

```python
table = df
```

```python
table["cluster"] = cluster
```

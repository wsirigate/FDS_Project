# Import Libary

```python
import pandas as pd
import numpy as np
```

# Import Data

```python
raw = pd.read_excel("หนี้สินเฉลี่ยต่อครัวเรือน จำแนกตามวัตถุประสงค์ของการกู้ยืม เป็นรายจังหวัด_พ.ศ._2545-2560_edit.xlsx")
```

```python
raw.head(10)
```

# Select column

```python
select1 = raw.iloc[:, 0:3]
select2 = raw.iloc[:, 11:12]
```

# Create key and add to column for merge

```python
lst = []
for i in range(539):
    lst.append(i)
```

```python
select1["key"] = lst
select2["key"] = lst
```

```python
select1.head()
```

```python
select2.head()
```

# Merge column

```python
raw_filter = select1.merge(select2, left_on='key', right_on='key')
```

## Delete key column

```python
del raw_filter['key']
```

# Dept in 2560

```python
raw_filter.head(10)
```

```python
df = raw_filter
```

```python
count = 0
for i in df.index:
    num = i-count
    if df.iloc[num]["วัตถุประสงค์ของการกู้ยืม"] == "หนี้สินทั้งสิ้น":
        df = df.drop([i])
        count += 1
df.head(10)
```
